### Fill the missing values by using (a)mean,(b)median,(c)mode

#### (a) fill null values by mean

```python
d.head(2)
```

```python
df= d.fillna(d.mean())
# df= d.fillna(d.mean(),inplace=True) for permanent changes
```

```python
df.head()
```

#### (b) fill null values by median

```python
d.head(2)
```

```python
df1= d.fillna(d.median())
# df1= d.fillna(d.median(),inplace=True) for permanent changes
```

```python
df1
```

#### (c) fill null values by mode

```python
d.head(2)
```

```python
df2= d.fillna(d.mode())
# df2= d.fillna(d.mode(),inplace=True) for permanent changes
```

```python
df2
```
